```python
#Dependencies
import pandas as pd
import numpy as np
```

```python
#csv file that contains the names of all chicago streets
csv_path = "Chicago_Street_Names.csv"
```

```python
#open the csv file with pandas and print out
street_df = pd.read_csv(csv_path, encoding = "utf-8")
street_df.head()
```

```python
#print tthe names of the columns
## 'Street ' is the column we want to sort
### Note that there is an extra space in 'Street '
list(street_df)
```

```python
#Sort out and alphabetize all street names in chicago that contain the suffix "dale"
```

```python
dales_df = street_df[street_df["Street "].str.contains("DALE", na = False)].sort_values("Street ", ascending = True)
total_dales = dales_df["Street "].unique()
total_dales
```
